```python
grafo = {
    1: [1, 2, 3],
    2: [1, 3],
    3: [1, 2, 4],
    4: [3]
}

print("Grafo:")
for nodo, adyacentes in grafo.items():
    print(nodo, "->", adyacentes)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

G.add_nodes_from(grafo.keys())

for nodo, adyacentes in grafo.items():
    for adyacente in adyacentes:
        G.add_edge(nodo, adyacente)

nx.draw(G, with_labels=True, node_color='skyblue', node_size=800, font_size=12, font_weight='bold')
plt.title("Grafo")
plt.show()
```

```python
num_nodos = len(grafo)
matriz_adyacencia = [[0] * num_nodos for _ in range(num_nodos)]

for nodo, adyacentes in grafo.items():
    for adyacente in adyacentes:
        matriz_adyacencia[nodo - 1][adyacente - 1] = 1

print("Matriz de Adyacencia:")
for fila in matriz_adyacencia:
    print(fila)
```

```python
parejas_organizadas = []

for nodo, adyacentes in grafo.items():
    for adyacente in adyacentes:
        if nodo < adyacente:
            parejas_organizadas.append((nodo, adyacente))

print("Parejas de nodos organizadas:")
for pareja in parejas_organizadas:
    print(pareja)
```

```python
num_nodos = len(grafo)
num_aristas = sum(len(adyacentes) for adyacentes in grafo.values()) // 2
matriz_incidencia = [[0] * num_aristas for _ in range(num_nodos)]

arista_idx = 0
for nodo, adyacentes in grafo.items():
    for adyacente in adyacentes:
        if nodo < adyacente:
            matriz_incidencia[nodo - 1][arista_idx] = 1
            matriz_incidencia[adyacente - 1][arista_idx] = -1
            arista_idx += 1

print("Matriz de Incidencia:")
for fila in matriz_incidencia:
    print(fila)
```

```python
parejas_organizadas = []

for nodo, adyacentes in grafo.items():
    for adyacente in adyacentes:
        parejas_organizadas.append((nodo, adyacente))

print("Parejas de nodos organizadas:")
for pareja in parejas_organizadas:
    print(pareja)
```

```python
grafo = {
    1: [1, 2, 3],
    2: [1, 3],
    3: [1, 2],
    4: [5],
    5: [1],
    6: [8]
}

import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

G.add_nodes_from(grafo.keys())

for nodo, adyacentes in grafo.items():
    for adyacente in adyacentes:
        G.add_edge(nodo, adyacente)

nx.draw(G, with_labels=True, node_color='skyblue', node_size=800, font_size=12, font_weight='bold')
plt.title("Grafo")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G1 = nx.Graph()
G2 = nx.Graph()

G1.add_edges_from([(1, 2), (2, 3), (3, 1)])

G2.add_edges_from([(4, 5), (5, 6), (6, 4)])

is_isomorphic = nx.is_isomorphic(G1, G2)
print(f"Los grafos G1 y G2 son isomorfos?: {is_isomorphic}")

plt.figure(figsize=(10, 5))

plt.subplot(121)
nx.draw(G1, with_labels=True, node_color="lightblue", font_weight="bold")
plt.title("Grafo G1")

plt.subplot(122)
nx.draw(G2, with_labels=True, node_color="lightgreen", font_weight="bold")
plt.title("Grafo G2")

plt.show()
```
