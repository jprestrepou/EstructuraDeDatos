```python
import csv
```

# ***Escritura en Archivos CSV:***

```python
datos = [
    ['Nombre', 'Edad', 'Correo'],
    ['Juan', '25', 'juan@XXXX.com'],
    ['María', '30', 'maria@XXXX.com']
]

with open('archivo.csv', 'w', newline='') as archivo_csv:
    escritor_csv = csv.writer(archivo_csv)
    escritor_csv.writerows(datos)
```

# ***Lectura de Archivos CSV***



```python
with open('archivo.csv', 'r') as archivo_csv:
    lector_csv = csv.reader(archivo_csv)
    for fila in lector_csv:
        print(fila)
```

# ***Especificar el Delimitador***

```python
with open('archivo.csv', 'r') as archivo_csv:
    lector_csv = csv.reader(archivo_csv, delimiter=';')
    for fila in lector_csv:
        print(fila)
```

# ***Manejo de Encabezados***

```python
with open('archivo.csv', 'r') as archivo_csv:
    lector_csv = csv.DictReader(archivo_csv)
    for fila in lector_csv:
        print(fila['Nombre'], fila['Edad'], fila['Correo'])
```

# ***Escribir con Encabezados***

```python
datos = [
    {'Nombre': 'Juan', 'Edad': '25', 'Correo': 'juan@example.com'},
    {'Nombre': 'María', 'Edad': '30', 'Correo': 'maria@example.com'}
]

with open('nuevo_archivo.csv', 'w', newline='') as archivo_csv:
    fieldnames = ['Nombre', 'Edad', 'Correo']
    escritor_csv = csv.DictWriter(archivo_csv, fieldnames=fieldnames)
    escritor_csv.writeheader()
    escritor_csv.writerows(datos)
```

# ***Personalizar el Formato de Escritura***

```python
import csv

datos = [
    ['Nombre', 'Edad', 'Correo'],
    ['Juan', '25', 'juan@example.com'],
    ['María', '30', 'maria@example.com']
]

with open('nuevo_archivo.csv', 'w', newline='') as archivo_csv:
    escritor_csv = csv.writer(archivo_csv, delimiter=';', quotechar='"', quoting=csv.QUOTE_MINIMAL)

    escritor_csv.writerows(datos)
```

# ***Especificar el Delimitador***




```python
with open('nuevo_archivo.csv', 'r') as archivo_csv:
    lector_csv = csv.reader(archivo_csv, delimiter=';')
    for fila in lector_csv:
        print(fila)
```

```python
import pandas as pd

datos = {
    'Nombre': ['Juan', 'María', 'Carlos', 'Laura', 'Pedro', 'Sofía', 'Andrés', 'Ana', 'Luis', 'Elena'],
    'Edad': [25, 30, 28, 35, 40, 22, 33, 27, 31, 36],
    'Correo': ['juan@example.com', 'maria@example.com', 'carlos@example.com', 'laura@example.com', 'pedro@example.com',
               'sofia@example.com', 'andres@example.com', 'ana@example.com', 'luis@example.com', 'elena@example.com']
}


df = pd.DataFrame(datos)
df.to_csv('datos_personales.csv', index=False, sep=';', quoting=1)
```

```python
df = pd.read_csv('datos_personales.csv', sep=';', quoting=1)

print("\nInformación del DataFrame:")
print(df.info())
```

```python
print("\nEstadísticas descriptivas del DataFrame:")
print(df.describe())
```

```python
print("\nPersonas menores de 30 años:")
print(df[df['Edad'] < 30])
```

```python
print("\nConteo de valores únicos en la columna 'Edad':")
print(df['Edad'].value_counts())
```

```python
import matplotlib.pyplot as plt

edades = datos['Edad']
nombres = datos['Nombre']

plt.figure(figsize=(10, 6))
plt.bar(nombres, edades, color='skyblue')
plt.title('Edades de las personas')
plt.xlabel('Nombre')
plt.ylabel('Edad')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
edades = datos['Edad']
correos = datos['Correo']

plt.figure(figsize=(10, 6))
plt.scatter(correos, edades, color='orange', alpha=0.7)
plt.title('Edad vs. Correo electrónico')
plt.xlabel('Correo electrónico')
plt.ylabel('Edad')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(edades, labels=nombres, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title('Distribución de edades')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
